```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import matplotlib.patches as mpatches
import seaborn as sns
import numpy as numpy
import pandas as pd
```

```python
df = pd.DataFrame({
    'Group': ['A', 'B', 'C', 'D', 'E'],
    'Value': [1, 5, 4, 3, 9]
})
```

```python
plt.barh(y = df.Group, width = df.Value)
```

```python
plt.barh(y = df.Group, 
    width = df.Value.sort_values(ascending=True), 
    color=['magenta', 'cyan'],
    edgecolor='black',
    linewidth = 3)
```

```python
plt.barh(y = df.Group, 
    width = df.Value.sort_values(ascending=True), 
    color = mcolors.TABLEAU_COLORS,
    edgecolor='black',
    linewidth = 3)
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.info()
```

```python
sns.set(style="darkgrid")
my_plot= sns.barplot(
    x = "total_bill",
    y = "day",   
    data=tips,
    color  ='cyan'
)
# plt.bar_label(container=my_plot, padding=10)
plt.figure(figsize=(10,6))
```

```python
sns.set(style="darkgrid")
sns.barplot(
    x = "total_bill",
    y = "day",   
    data=tips,
    color ='cyan',
    hue='smoker'
)
plt.figure(figsize=(10,6))
```

```python
tips.head()
```

```python
total = tips.groupby('day')['total_bill'].sum().reset_index()
total
```

```python
bar1 = sns.barplot(x='day', y='total_bill', data=total, color = 'darkblue')
```

```python
smoker = tips[tips.smoker=='Yes']
```

```python
bar2 = sns.barplot(x='day', 
                   y='total_bill', 
                   data=smoker, 
                   estimator = sum, 
                   errorbar=None, 
                   color='lightblue')
```

```python
bar1 = sns.barplot(x='day', y='total_bill', data=total, color = 'darkblue')
bar2 = sns.barplot(x='day', 
                   y='total_bill', 
                   data=smoker, 
                   estimator = sum, 
                   errorbar=None, 
                   color='lightblue')

top_bar = mpatches.Patch(color='darkblue', label='non-smoker')
bottom_bar = mpatches.Patch(color='lightblue', label='smoker = Yes')
plt.legend(handles = [top_bar, bottom_bar])
plt.show()
```
