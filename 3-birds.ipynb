```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import matplotlib.patches as mpatches
import seaborn as sns
import numpy as numpy
import pandas as pd
```

```python
birds_url = "data/birds.csv"
```

```python
birds=pd.read_csv(birds_url)
```

```python
birds.info()
```

```python
birds.head()
```

```python
birds.Family.value_counts()
```

```python
# Family - Family value_counts

family_counts = birds["Family"].value_counts()
# Seri -> Dataframe
family_counts_df = pd.DataFrame({'Family': family_counts.index, 'SpeciesCount': family_counts.values})

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x='Family', y='SpeciesCount', data=family_counts_df)
plt.xticks(rotation=90)
plt.xlabel('Family')
plt.ylabel('Number of Species')
plt.title('Number of Species per Family')
plt.show()
```

```python
plt.barh(y = birds["Family"].value_counts().sort_values(ascending=False).iloc[0:5].index, 
         width =birds["Family"].value_counts().sort_values(ascending=False).iloc[0:5].values)
```

```python
sns.histplot(data=birds, x='MinBodyMass', kde=True)
```

```python
sns.histplot(data=birds, x='MaxWingspan', kde=True)
```

```python
#####cinsine göe max kanat uzunlukları ########
genus_max_wingspan_df = birds.groupby('Genus')['MaxWingspan'].max().reset_index()
top5_genus_max_wingspan_df = genus_max_wingspan_df.head()
# Plotting
plt.figure(figsize=(6, 3))
sns.barplot(x='Genus', y='MaxWingspan', data=top5_genus_max_wingspan_df)
plt.xticks(rotation=90)
plt.xlabel('Kuş Cinsi')
plt.ylabel('Maksimum Kanat Uzunluğu')
plt.title('Kuş Cinsine Göre Maksimum Kanat Uzunluğu')
plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.boxplot(x = birds['Family'],
            y = birds['MinBodyMass']
            )

plt.xticks(rotation=90)
plt.xlabel('Family')
plt.ylabel('MinBodyMass')
plt.title('MinBodyMass per Family')
plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.boxplot(x = birds['Family'],
            y = birds['ConservationStatus']
            )

plt.xticks(rotation=90)
plt.xlabel('Family')
plt.ylabel('')
plt.title('ConservationStatus per Family')
plt.show()
```
